<h3>3. Researching </h3> 
Retrieve the highest correlated and cointegrated pairs.

```python
best_coins = Researcher()
best_coins.load_data()
# best_coins.update_data(timeframe='1d', interval='1 years ago', min_correlation=0.83) # uncomment to update data (may take some time)
best_coins.researched_df
```

<h3>4. Backtesting </h3> 
Get results for each pairs combination.

```python
backtest = Backtester()
backtest.read_data(best_coins)
backtest.run_backtest()
backtest.backtested_df
```
